# Jupyter Notebook — Setup Guide (Windows)

Step-by-step : ouvrir un terminal, créer un environnement Python, lancer Jupyter, installer les librairies.


---
## 1. Ouvrir un terminal

**Méthode rapide :** `Win + R` → taper `cmd` → Entrée

**Méthode alternative :** cliquer sur la barre de recherche Windows (en bas à gauche) → taper `cmd` ou `Command Prompt` → cliquer dessus

**Résultat attendu :** une fenêtre noire s'ouvre avec un curseur clignotant :
```
C:\Users\ton_nom>
```

> **Si ça ne marche pas :**
> - Si `cmd` n'est pas reconnu → essayer `Win + R` → taper `powershell` → Entrée (PowerShell accepte les mêmes commandes)
> - Si la fenêtre s'ouvre puis se ferme immédiatement → clic droit sur `Command Prompt` → "Run as administrator"


---
## 2. Aller dans le dossier de travail

```bat
cd C:\Users\ton_nom\Desktop\mon_projet
```

**Résultat attendu :** le prompt change :
```
C:\Users\ton_nom\Desktop\mon_projet>
```

> **Si ça ne marche pas :**
> - `The system cannot find the path specified` → le dossier n'existe pas, le créer d'abord :
>   ```bat
>   mkdir C:\Users\ton_nom\Desktop\mon_projet
>   cd C:\Users\ton_nom\Desktop\mon_projet
>   ```
> - Tu ne connais pas le chemin → ouvrir le dossier dans l'explorateur, cliquer dans la barre d'adresse en haut, copier le chemin, coller dans le terminal
> - Pour changer de disque (ex: passer sur `D:\`) → taper `D:` puis Entrée, ensuite `cd D:\mon_dossier`


---
## 3. Vérifier que Python est installé

```bat
python --version
```

**Résultat attendu :**
```
Python 3.x.x
```

> **Si ça ne marche pas :**
> - `'python' is not recognized` → essayer `python3 --version`. Si ça marche → utiliser `python3` au lieu de `python` pour toutes les commandes suivantes
> - Ni `python` ni `python3` ne marchent → Python n'est pas installé ou pas dans le PATH. Installer depuis [python.org](https://python.org) en **cochant "Add Python to PATH"**
> - Affiche `Python 2.x.x` → installer Python 3 depuis python.org


---
## 4. Créer un environnement virtuel

```bat
python -m venv venv
```

**Résultat attendu :** rien ne s'affiche (silence = succès). Vérifier avec :
```bat
dir venv
```
→ doit lister `Include`, `Lib`, `Scripts`

> **Si ça ne marche pas :**
> - `No module named venv` → alternative :
>   ```bat
>   pip install virtualenv
>   virtualenv venv
>   ```
> - `Permission denied` → relancer le terminal en administrateur (clic droit → "Run as administrator")
> - Le dossier `venv` existe déjà → le supprimer puis recréer :
>   ```bat
>   rmdir /s /q venv
>   python -m venv venv
>   ```


---
## 5. Activer le venv

**Dans cmd :**
```bat
venv\Scripts\activate
```

**Résultat attendu :** `(venv)` apparaît au début du prompt :
```
(venv) C:\Users\ton_nom\Desktop\mon_projet>
```

> **Si ça ne marche pas :**
> - `'venv\Scripts\activate' is not recognized` → vérifier que tu es dans le bon dossier (celui qui contient le dossier `venv`). Taper `dir` pour voir les fichiers.
> - **Si tu es dans PowerShell** (pas cmd) → la commande est :
>   ```powershell
>   venv\Scripts\Activate.ps1
>   ```
> - **PowerShell bloque avec "execution of scripts is disabled"** → exécuter d'abord :
>   ```powershell
>   Set-ExecutionPolicy -Scope CurrentUser -ExecutionPolicy RemoteSigned
>   ```
>   puis relancer `venv\Scripts\Activate.ps1`
>
> Pour désactiver plus tard : `deactivate`


---
## 6. Installer et lancer Jupyter

```bat
pip install notebook
jupyter notebook
```

**Résultat attendu :**
1. `pip install` → lignes de téléchargement, finit par `Successfully installed ...`
2. `jupyter notebook` → le terminal affiche :
```
[I 09:15:00] Jupyter Notebook is running at:
http://localhost:8888/?token=abc123...
```
Le navigateur s'ouvre automatiquement sur l'interface Jupyter.

> **Si ça ne marche pas :**
> - `pip` non reconnu → `python -m pip install notebook`
> - `jupyter` non reconnu → `python -m jupyter notebook`
> - Le navigateur ne s'ouvre pas → copier-coller l'URL `http://localhost:8888/?token=...` du terminal dans le navigateur manuellement
> - `port 8888 is already in use` → `jupyter notebook --port 8889`

**Créer un notebook :** dans l'interface Jupyter → bouton **New** (en haut à droite) → **Python 3**


---
## 7. Installer les librairies (dans le notebook)

Exécuter les cellules ci-dessous **une fois** dans le notebook. Le `!` exécute des commandes terminal depuis Jupyter.


````python
# ── Vérifier ce qui est déjà installé ──
import numpy, pandas, sklearn
print(f"numpy {numpy.__version__}  |  pandas {pandas.__version__}  |  sklearn {sklearn.__version__}")
print("OK — ces 3 sont déjà disponibles")
````

````python
# ── Installer le reste ──

# Stats & econometrics
!pip install --quiet statsmodels scipy

# Visualization
!pip install --quiet matplotlib seaborn

# Gradient boosting
!pip install --quiet xgboost lightgbm

# Interpretability
!pip install --quiet shap

# Excel file support
!pip install --quiet openpyxl

# Progress bars
!pip install --quiet tqdm

print("\nDone")
````

**Résultat attendu :** pour chaque ligne, soit `Requirement already satisfied` (déjà installé) soit `Successfully installed ...`

> **Si ça ne marche pas :**
> - `ERROR: Could not install packages due to an EnvironmentError: [Errno 13]` → le venv n'est pas activé. Retourner au terminal et refaire l'étape 5.
> - `ERROR: No matching distribution found for xgboost` → la version Python est trop ancienne ou le package n'est pas dispo pour l'OS. Pas grave, sklearn couvre 90% des besoins. Continuer sans.
> - Très lent / timeout → essayer avec un miroir plus rapide :
>   ```
>   !pip install --quiet --index-url https://pypi.org/simple/ statsmodels
>   ```
